**Libraries**

```python
# Selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
# Navigator option (chrome)
options =  webdriver.ChromeOptions()
options.add_argument('--start-normal')
options.add_argument('--disable-extensions')
prefs = {"profile.default_content_setting_values.notifications" : 2}
options.add_experimental_option("prefs",prefs)
driver_path = 'C:\\Users\\Jazmine\\Downloads\\Applications\\WebDriver\\chromedriver.exe'

# add options to webdriver
driver = webdriver.Chrome(driver_path, chrome_options=options)

# Source website (Open navigator)
driver.get('https://cryptowat.ch/es-es/')

# Click button and search into textbox
WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.CSS_SELECTOR,
                                      'ul.flex-none')))\
    .click()

time.sleep(5)

# Search element by xpath    
texto_columnas = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[2]/div[1]/div/div[2]/div[2]/div[2]/div/div[1]/div/converse-root/div/div[1]/div/div/div[2]/div[1]/div[1]/div[1]/converse-chat-content/converse-message-history')

# Save the result on HTML format
element_html = texto_columnas.get_attribute("outerHTML")

# Print result
print(element_html)

# Close navigator chrome
driver.quit()
```

# BeautifulSoup

**Filter bye HTML format**

```python
# Convert BeautifulSoup format
soup = BeautifulSoup(element_html)

# Prettify the html
prettyHTML = soup.prettify()   

print(prettyHTML)
```

```python
news_content = soup.findAll({"converse-chat-message"})
print(news_content)
```

Save data into Datataframe

```python
# Prepare dataframe for save data into csv
df_news = pd.DataFrame(columns=("time", "user", "message"))

for i in range(len(news_content)):
    try:
        # Print date
        date = news_content[i].find("time")
        print("Date:", date['timestamp'])
        date = date['timestamp']

        username = news_content[i].find("span", {"class": "chat-msg__author"})
        print("UserName:",username.text)
        username = username.text

        message_text = news_content[i].find("converse-chat-message-body", {"class": "chat-msg__text"})
        print("Message:", message_text.text)
        message_text = message_text.text

        print('---------------------------------------------------------------------')
        
        df_news.loc[len(df_news)] = [date, username, message_text]
        
    except Exception as e:
            pass
```
